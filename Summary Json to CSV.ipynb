**Making empty Lists**

```python
files = []
start_date = []
end_date = []
city = []
venue = []
event_name = []
match_type = []
match_type_number = []
team_type = []
team_1 = []
team_2 = []
toss_win = []
toss_decision = []
team_1_runs = []
team_2_wickets = []
team_2_runs = []
team_1_wickets = []
outcome = []
```

**Function for total runs in first and second innings**

```python
def total_runs(team):
    i = 0
    runs = []
    
    for over in data['innings'][team]['overs']:
        for ball in data['innings'][team]['overs'][i]['deliveries']:
            # total runs made in this over
            runs.append(ball['runs']['total'])
        i = i + 1
    return sum(runs)
```

**Function for Total wickets taken by team**

```python
def total_wickets(team):
    i = 0
    wickets = []
    for over in data['innings'][team]['overs']:
        for ball in data['innings'][team]['overs'][i]['deliveries']:
            # Checking if wicket is taken or not in this over
            for x in ball.keys():
                if x == 'wickets':
                    wickets.append(1)
        i = i + 1
    return len(wickets)
```

**Looping through all the json files and collecting necessary data**

```python
for file in os.listdir("Data/woman/2022_female_json"):
    if file.endswith(".json"):
        files.append(os.path.join("Data/woman/2022_female_json", file))

for file in files:
    with open(file) as f: 
        data = json.load(f)
        print(f'fileName: {file}')
        
        if ('dates' or 'date') in data['info'].keys():
            # start date 
            start_date.append(data['info']['dates'][0])

            # end_date
            end_date.append(data['info']['dates'][-1])
        else:
            start_date.append('missing')
            end_date.append("missing")
            
        # city
        if 'city' in data['info'].keys():
            city.append(data['info']['city'])
        else:
            city.append('missing')
        # venue
        if 'venue' in data['info'].keys():
            venue.append(data['info']['venue'])
        else:
            venue.append('missing')
        
        # event_name
        if 'event' in data['info'].keys():
            event_name.append(data['info']['event']['name'])
        else:
            event_name.append('missing')
            
        # match type
        if 'match_type' in data['info'].keys():
            match_type.append(data['info']['match_type'])
        else:
            match_type.append('missing')
        
        # match_type_number
        if 'match_type_number' in data['info'].keys(): 
            match_type_number.append(data['info']['match_type_number'])
        else:
            match_type_number.append('missing')
            
        # team_type
        if 'team_type' in data['info'].keys():
            team_type.append(data['info']['team_type'])
        else:
            team_type.append('missing')
            
        if 'teams' in data['info'].keys():
            # team 1
            team_1.append(data['info']['teams'][0])

            # team 2
            team_2.append(data['info']['teams'][1])
        else:
            team_1.append('missing')

            team_2.append('missing')

        # toss_winner
        toss_win.append(data['info']['toss']['winner'])

        # toss_decision
        toss_decision.append(data['info']['toss']['decision'])

        # OutCome (who won ?)
        if 'winner' in data['info']['outcome'].keys():
            outcome.append(f"{data['info']['outcome']['winner']} won")
        else:
            outcome.append(data['info']['outcome']['result'])

        # team_1 & team 2 runs and wickets
        
        # if team 1 won toss but decieds to field first then they
        # will be playing bat in second inning and team 2 playing first bat
        # while calculating total runs in inning 1 default goes to team 1 and
        # 2 inning runs to 2 team but it can change as I explained in first two lines
        # To resolve that it goes thorugh this next if else statments.
        
        if (data['info']['toss']['winner'] ==  data['info']['teams'][0] and data['info']['toss']['decision'] =='bat') or (data['info']['toss']['winner'] == data['info']['teams'][1] and data['info']['toss']['decision'] =='field') :
            print(f"team_1_runs: {total_runs(0)} ")
            team_1_runs.append(total_runs(0))
            team_2_wickets.append(total_wickets(0))
            if 2 <= len(data['innings']): 
                team_2_runs.append(total_runs(1))
                team_1_wickets.append(total_wickets(1))
            else:
                team_2_runs.append('missing')
                team_1_wickets.append('missing')
    
        if (data['info']['toss']['winner'] == data['info']['teams'][1] and data['info']['toss']['decision'] =='bat') or (data['info']['toss']['winner'] == data['info']['teams'][0] and data['info']['toss']['decision'] =='field'):
            print(f"team_2_runs: {total_runs(0)} ")
            team_2_runs.append(total_runs(0))
            team_1_wickets.append(total_wickets(0))
            if 2 <= len(data['innings']): 
                team_1_runs.append(total_runs(1))
                team_2_wickets.append(total_wickets(1))
            else:
                team_1_runs.append('missing')
                team_2_wickets.append("missing")
            
```

```python
df = pd.DataFrame(columns = ['start_date', 'end_date',
                             'city','venue',
                              'event_name', 'match_type',
                            'match_type_number', 'team_type',
                            'team_1', 'team_2', 'toss_win',
                            'toss_decision', 'team_1_runs',
                            'team_2_runs', 'team_1_wickets', 'team_2_wickets',
                            'OutCome'])
```

```python
df['start_date'] = start_date
df['end_date'] = end_date
df['city'] = city
df['venue'] = venue
df['event_name'] = event_name
df['match_type'] = match_type
df['match_type_number'] = match_type_number
df['team_type'] = team_type
df['team_1'] = team_1
df['team_2'] = team_2
df['toss_win'] = toss_win
df['toss_decision'] = toss_decision
df['team_1_runs'] = team_1_runs
df['team_2_runs'] = team_2_runs
df['team_1_wickets'] = team_1_wickets
df['team_2_wickets'] = team_2_wickets
df['OutCome'] = outcome
```

```python
df.to_csv('Clean/woman/Summarized/2022_female.csv', index=False)
```
