# IPL all Matches ball to ball data

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Data/ipl_all_matches.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.isna().sum()
```

```python
# Replacing missing value in wides, noballs, byes, legbyes, penalty column with 0

df['wides'] = df['wides'].fillna(0)
df['noballs'] = df['noballs'].fillna(0)
df['byes'] = df['byes'].fillna(0)
df['legbyes'] = df['legbyes'].fillna(0)
df['penalty'] = df['penalty'].fillna(0)
df['other_wicket_type'] = df['other_wicket_type'].fillna(0)
df['other_player_dismissed'] = df['other_player_dismissed'].fillna(0)
```

```python
# Replacing missing value in column wicket_type with not_out

df['wicket_type'] = df['wicket_type'].fillna('Not Out')
df['player_dismissed'] = df['player_dismissed'].fillna('No Name')
```

```python
df.head()
```

```python
# Type Casting

df.dtypes
```

```python
df = df.replace('2007/08', '2008')
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])  
```

```python
df = df.astype({'wides':int, 'noballs': int, 'byes': int, 'legbyes': int, 'penalty': int, 'other_wicket_type': int, 'other_player_dismissed': int})
```

```python
df.head()
```

```python
df.set_index('match_id')
```

```python
df.to_csv('Data/Clean/ipl_all_matches_dataset.csv', index=False)
```
